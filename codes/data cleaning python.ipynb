```python
# read data from csv file
import pandas as pd
df = pd.read_csv('C:/Users/USER-PC/Desktop/Trump_fb_statuses.csv')
print(df.head())
```

```python
# see whether there are some na value
df.isna().sum()
```

```python
# drop useless columns named 'status_link' and 'status_id' and 'link_name'
df.drop(['status_link', 'status_id', 'link_name'], axis=1, inplace=True)
print(df.head())
```

```python
# drop na value in column 'status_message'  
df.dropna(subset=['status_message'], inplace=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

# INITIALIZE COUNT VECTORIZER
count_vectorizer = CountVectorizer()    

# RUN COUNT VECTORIZER ON   'status_message' COLUMN
count_data = count_vectorizer.fit_transform(df['status_message'])
print(type(count_data))
```

```python
# VOCABULARY DICTIONARY
print("vocabulary = ",count_vectorizer.vocabulary_)   

# STOP WORDS 
print("stop words =", count_vectorizer.stop_words)

# col_names
col_names=count_vectorizer.get_feature_names()
print("col_names=",col_names)

# CREATE DATAFRAME
df_count = pd.DataFrame(count_data, columns=count_vectorizer.vocabulary_)

# SUM WORD COUNTS
word_counts = df_count.sum(axis=0)

# SORT WORD COUNTS
word_counts = word_counts.sort_values(ascending=False)

# PRINT TOP 20 WORDS
print(word_counts.head(20))
```

```python
# write data to csv file
word_counts.to_csv('C:/Users/USER-PC/Desktop/word_counts.csv')
```
